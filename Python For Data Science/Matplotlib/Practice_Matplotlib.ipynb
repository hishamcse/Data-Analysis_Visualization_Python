```
import matplotlib.pyplot as plt
```

```
%matplotlib inline
```

```
import numpy as np
x = np.linspace(0,5,11)
y = x**2
print(x)
print(y)
```

```
# functional method
plt.plot(x,y)
```

```
plt.plot(x,y,'g-')
plt.xlabel('time')
plt.ylabel('distance')
plt.title('distance vs time graph')
```

```
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

```
# object oriented method (recommended)
fig = plt.figure()

axes = fig.add_axes([.1,.1,.8,.8])

axes.plot(x,y,'bx-')
axes.set_xlabel('time')
axes.set_ylabel('distance')
axes.set_title('distance vs time graph')
```

```
fig = plt.figure()

axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([.2,.5,.4,.3])

axes1.plot(x,y,'g--')
axes1.set_xlabel('time')
axes1.set_ylabel('distance')
axes1.set_title('distance vs time graph')

axes2.plot(y,x,'r*-')
axes2.set_xlabel('distance')
axes2.set_ylabel('time')
axes2.set_title('time vs distance graph')
```

```
fig,axes = plt.subplots(3,3)

plt.tight_layout()     # to make enough spaces between the figures
```

```
fig,axes = plt.subplots(3,1)

for cur_ax in axes:
    cur_ax.plot(x,y)

plt.tight_layout()
```

```
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y,color='purple',linewidth=.5)
# ax.plot(x,y,color='#ff00ff',lw=5,alpha=1.5,linestyle='steps')      # linewidth=lw  linestyle= ls = ':' or '--' or '-.' or 'steps'
# ax.plot(x,y,'r',lw=1,ls='-',marker='+',markersize=15)                  # marker = 'o' or '*' or '+' or 'x' or '1' or '2' etc..

ax.plot(x,y,'r',lw=1,ls='-',marker='o',markersize=15,
        markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='purple')
```

```
fig,axes=plt.subplots(1,2)

axes[0].plot(x,y,'g',lw=1.5,ls='--')
axes[1].plot(y,x,'b',lw=2,ls='-.')

axes[0].set_xlim([0,2])
axes[0].set_ylim([0,1])

plt.tight_layout()
```

***special type graph.***


```
np.random.seed(0)

x, y = np.random.randn(2, 100)
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax1.xcorr(x, y, usevlines=True, maxlags=50, normed=True, lw=2)
ax1.grid(True)
ax1.axhline(0, color='black', lw=2)

ax2 = fig.add_subplot(212, sharex=ax1)
ax2.acorr(x, usevlines=True, normed=True, maxlags=50, lw=2)
ax2.grid(True)
ax2.axhline(0, color='black', lw=2)
```
